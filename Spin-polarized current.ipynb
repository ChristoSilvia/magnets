Now that I have [mumax3](mumax.github.io) working, I'm going to try to get some data from it. I'm diving in to playing around with the example [Slonczewski STT](http://mumax.github.io/examples.html) (scroll midway).

This example is a 100nmx50nmx5nm Py ellipsoid, initially magnetized to the right.  Some current is pulsed at it, and its behavior changes.  I'm not sure why.

Here is the code:
```
// geometry
sizex := 160e-9
sizey := 80e-9
sizez := 5e-9

nx := 256
ny := 128

setgridsize(nx, ny, 1)
setcellsize(sizex/nx, sizey/ny, sizez)
setgeom(ellipse(sizex,sizey))

// setup free layer
Msat  = 800e3
Aex   = 13e-12
alpha = 0.01
m     = uniform(1, 0, 0)

// setup spacer layer parameters
lambda       = 1
Pol          = 0.5669
epsilonprime = 0

// setup fixed layer polarization
angle := 20
px := cos(angle * pi/180)
py := sin(angle * pi/180)
fixedlayer = vector(px, py, 0)

// send current
Jtot := -0.004 //Amps
area := sizex*sizey*pi/4
jc   := Jtot / area //Amps/m^2 current density
J = vector(0,0,jc)

// schedule, output, and run
autosave(m, 5e-12)
tableautosave(5e-12)
run(1e-9)
```



````python
%pylab inline
import numpy as np

def plot_magnetization(table_filename,x=True,y=True,z=True,prefix=""):
  data = np.loadtxt(table_filename).T
  pylab.axis([0,max(data[0]),-1.1,1.1])
  if x:
    pylab.plot(data[0],data[1],label=prefix+"m_x")
  if y:
    pylab.plot(data[0],data[2],label=prefix+"m_y")
  if z:
    pylab.plot(data[0],data[3],label=prefix+"m_z")
  pylab.legend(loc=3)
````

It produces this output when plotted:

````python

plot_magnetization("initial/layers_many/table_008.txt", y=False, z=False)
````

````python
pylab.title("Dynamic Responses to Spin-Polarized Current Pulses \nin 100nmx50nmx5nm Py ellipsoid")
pylab.xlabel("Time (sec)")
pylab.ylabel("<m_x>")

plot_magnetization("initial/layers_many/table_001.txt", y=False, z=False, prefix=".001 A")
plot_magnetization("initial/layers_many/table_004.txt", y=False, z=False, prefix=".004 A")
plot_magnetization("initial/layers_many/table_005.txt", y=False, z=False, prefix=".005 A")
plot_magnetization("initial/layers_many/table_006.txt", y=False, z=False, prefix=".006 A")
plot_magnetization("initial/layers_many/table_007.txt", y=False, z=False, prefix=".007 A")
plot_magnetization("initial/layers_many/table_008.txt", y=False, z=False, prefix=".008 A")
plot_magnetization("initial/layers_many/table_009.txt", y=False, z=False, prefix=".009 A")
plot_magnetization("initial/layers_many/table_01.txt", y=False, z=False,prefix=".01 A")
````
